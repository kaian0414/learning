```python
df3 = pd.DataFrame(np.array([
    [1, np.nan, 2],
    [np.nan, 3, 4],""
    [5, 6, np.nan],
    [7, 8, 9]
]), columns=['A', 'B', 'C'])
```

```python
# 檢查缺失
df3.isnull()
```

```python
# 有缺失值的row都del
df3.dropna()
```

```python
# 有缺失值的col都del
df3.dropna(axis='columns')
```

```python
# 缺失值填充 (自行設定值)
df3.fillna(value=5)
```

```python
# 缺失值填充 (mean值)
df3.fillna(value=df3.mean())
```

```python
def make_df(cols, idx):
  data = {c: [str(c)+str(i) for i in idx] for c in cols}
  print(data)
  return pd.DataFrame(data, idx)
```

```python
df4 = make_df('AB', [1, 2])
df5 = make_df('AB', [3, 4])
df6 = make_df('AB', [1, 2])
df7 = make_df('BC', [1, 2])
```

```python
# 垂直合拼
pd.concat([df4, df5])
```

```python
# 對齊合拼merge
pd.merge(df6, df7)
```

groupby

```python
df = pd.DataFrame({
    'key': ['A', 'B', 'A', 'A', 'B', 'C'],
    'data1': range(6),
    'data2': np.random.randint(0, 10, size=6)
})
```

```python
df
```

```python
# 用key分組
df.groupby('key')
```

```python
df.groupby('key').sum() # 找sum()
```

```python
df.groupby('key').mean() # 找mean()
```

```python
df.groupby('key')['data2'].sum() # 只取其中一列出來做sum()
```

```python
# 分組再找'min', 'median', 'max'
df.groupby('key').agg(['min', 'median', 'max'])
```
